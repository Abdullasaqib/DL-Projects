```python
# Reshape
## the lstm model accepts only 3d array.so we are reshaping it into 3d


####(number of samples,number of time stamp,no.of features)

# (1198,60,1)

X_train = np.reshape(X_train,(X_train.shape[0],X_train.shape[1],1))
```

```python
X_train.shape
```

```python
X_train.ndim
```

```python
# Model buiding

# imp algo
from keras.models import Sequential
```

```python
from keras.layers import Dense
from keras.layers import LSTM  # main layer
from keras.layers import Dropout  #  to avoid overfitting
```

```python
# Initialize the model
stock_model = Sequential()
```

```python
# Add the layers

# first layer:  input + hidden (LSMT)

stock_model.add(LSTM(units=50,return_sequences=True,input_shape=(X_train.shape[1],1)))
stock_model.add(Dropout(0.2))

# second hidden (LSTM)
stock_model.add(LSTM(units=50,return_sequences=True))
stock_model.add(Dropout(0.2))

# third hidden (LSTM)
stock_model.add(LSTM(units=50,return_sequences=False))
stock_model.add(Dropout(0.2))

# last;  output layer

stock_model.add(Dense(units=1),)

```

```python
# Compile the Model
stock_model.compile(optimizer='adam',loss='mean_squared_error')
```

```python
# Train the model
stock_model.fit(X_train,y_train,epochs=1000,batch_size=32)
```

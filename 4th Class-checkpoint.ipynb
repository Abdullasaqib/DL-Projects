# Convolutional_neural_network

## Importing the necessary libraries

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
path = r"D:\Datasets\Deep Learning\Car-Bike-Dataset"
```

# Part 1 - Data Preprocessing
 

### Preprocessing the Training set 

```python
# image augmentation
train_datagen = ImageDataGenerator(rescale = 1./255, #This rescales the pixel values of the images from the range [0, 255] to [0, 1].
                                   shear_range = 0.2,
                                   zoom_range = 0.2)

training_set = train_datagen.flow_from_directory(path+'/Training_Set',
                                                 target_size = (64, 64), #resizes all images to 64x64 pixels.because most neural networks require all input images to have the same dimensions.
                                                 batch_size = 32,
                                                 class_mode = 'binary')
```

### Preprocessing the Test set# 

```python
test_datagen = ImageDataGenerator(rescale = 1./255)
test_set = test_datagen.flow_from_directory(path+'/Test_Set',
                                            target_size = (64, 64),
                                            batch_size = 32,
                                            class_mode = 'binary')
```

## Part 2 - Building the CNN

### Initialising the CNN layers for Feature Extraction

```python
# initialize

from keras.models  import Sequential
from keras.layers import Conv2D , MaxPool2D , Flatten , Dense
Image_Classifier = Sequential()

# Add Convolution layer
Image_Classifier.add(Conv2D(filters=32,kernel_size=3,activation='relu',input_shape=(64,64,3)))
# output --> Feature Map

# pooling ..
Image_Classifier.add(MaxPool2D(pool_size=2,strides=2))
# output --> pooled Feature Map
```

```python
# Second Conv. Layer
Image_Classifier.add(Conv2D(filters=32,kernel_size=3,activation='relu'))
Image_Classifier.add(MaxPool2D(pool_size=2,strides=2))
```

```python
# Flattening layer -- matrix (2d) to vector (1d)
Image_Classifier.add(Flatten())
```

### Building the fully connected network 

```python
# full connection
Image_Classifier.add(Dense(units=128,activation='relu'))

# output layer
Image_Classifier.add(Dense(units=1,activation='sigmoid'))
```

### Compiling and Training the model

```python
# Compile
Image_Classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])

# Train the model
Image_Classifier.fit(x=training_set,validation_data=test_set,epochs =10)
```

# Testing the image with a sample image

```python
### testing the model using a sample image

import numpy as np

from tensorflow.keras.preprocessing import image

## loading the test image

test_image = image.load_img(r"D:\Datasets\Deep Learning\Car-Bike-Dataset\train size\Bike\Bike (231).jpeg", target_size = (64, 64))

## Converting the Image to an Array

test_image = image.img_to_array(test_image)

# Expanding the Dimensions
##  adds an additional dimension to the array, making the shape (1, 64, 64, 3).
###This is necessary because Keras models expect a batch of images as input, even if it's just a single image.

test_image = np.expand_dims(test_image, axis = 0)

# Making a Prediction

result = Image_Classifier.predict(test_image)

# checking the predicted result
## result[0][0] refers to the first element of the first prediction in the batch,which should be the class label or probability.

if result[0][0] == 1:
  prediction = 'car'
else:
  prediction = 'bike'
print(prediction)
```
